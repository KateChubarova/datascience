```python
import pandas as pd

df = pd.read_csv('https://stepik.org/media/attachments/lesson/779916/3_1_3.csv')

# В переменной ans находится правильный ответ для df
ans = pd.read_csv('https://stepik.org/media/attachments/lesson/779916/3_1_3_ans.csv')
```

```python
df.groupby('stop_times').count()
```

```python
feature_2helper = lambda x: x.sort_values(ascending=False).iloc[1]
```

```python
feature_3helper = lambda x: min(x)/max(x)
```

```python
feature_4helper = lambda x: x.mode()
```

```python
feature_5helper = lambda x: x.std()
```

```python
df.groupby("car_id", as_index=False).agg(feature_1=('stop_times', 'nunique'),
                                        feature_2=('speed_max', feature_2helper),
                                        feature_3=('rating', feature_3helper),
                                        feature_4=('stop_times', feature_4helper),
                                        feature_5helper=('ride_cost', feature_5helper))
```

```python
df.groupby('car_id', as_index=False).agg(feature_4=('stop_times', feature_4helper))
```
