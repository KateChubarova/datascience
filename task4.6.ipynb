```python
import pandas as pd

df = pd.read_csv('stratified_valid.csv')
```

```python
df
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
N = 10 
```

```python
df['target_reg_group'] = pd.qcut(df['target_reg'], N, labels = [f'group_{i}' for i in range(N)])
```

```python
df
```

```python
skf = StratifiedKFold(n_splits=N, shuffle = True)
y = df['target_reg_group']
df['num_fold'] = None
```

```python
df['num_fold'] = pd.qcut(df['target_reg'], q=N, labels=range(N))
```

```python
df
```

```python
for fold, (train_indexes, test_indexes) in enumerate(skf.split(df, df['target_reg_group'])):
    df.loc[test_indexes, 'num_fold'] = fold
```
