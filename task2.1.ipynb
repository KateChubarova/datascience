```python
pd.crosstab(
    df["target_class"]
    , df["model"]
    , margins = True
    #, normalize = True
).style.background_gradient(cmap="RdYlGn")
```

```python
import matplotlib.pyplot as plt
```

```python
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Создание тепловой карты с отображением числовых значений
sns.heatmap(data, annot=True, cmap='YlGnBu', fmt='d')

plt.title('Пример тепловой карты с аннотациями')
plt.xlabel('X-метки')
plt.ylabel('Y-метки')

plt.show()
```
