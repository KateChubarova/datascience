Добавьте столбец flag, который равен 'good_line', если rating > 5 and speed_max < 100. 'bad_line', если  rating > 5 and speed_max > 100. 'undefined' в остальных случаях.

```python
def line(x):
    if x.rating>5 and x.speed_max<100:
        return 'good_line'
    if x.rating>5 and x.speed_max>10:
        return 'bad_line'
    else: return 'undefined'
```

```python
df['flag'] = df.apply(lambda x: line(x), axis=1)
```

Добавьте столбец speed_check, который принимает значения 'avg_more', если distance/ride_duration > speed_avg, иначе 'avg_less'.

```python
def speed(x):
    if x.distance/x.ride_duration > x.speed_avg:
        return 'avg_more'
    else: return 'avg_less'
```

```python
df['speed_check'] = df.apply(lambda x: speed(x), axis=1)
```

```python
df.drop(cols2drop, axis=1, inplace=True)
```
