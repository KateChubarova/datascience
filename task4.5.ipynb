```python
import pandas as pd

rides = pd.read_csv('https://stepik.org/media/attachments/lesson/779916/3_1_6.csv')
```

```python
transform_75 = lambda x: x.quantile(0.75)
```

```python
min_max_mean = lambda x: (max(x)-min(x))/x.mean()
```

```python
count_40 = lambda x: x[x>40].count()
```

```python
sum_40 = lambda x: x[x>40].sum()
```

```python
rides['feature_1'] = rides.groupby('user_id', as_index=False)['ride_cost'].transform(transform_75)
```

```python
rides['feature_2'] = rides.groupby('user_id', as_index=False)['rating'].transform(min_max_mean)
```

```python
rides['feature_3'] = rides.groupby('car_id', as_index=False)['ride_duration'].transform(count_40)
```

```python
rides['feature_4'] = rides.groupby('car_id', as_index=False)['ride_duration'].transform(sum_40)
```
