```python
import pandas as pd
df = pd.read_csv('https://stepik.org/media/attachments/lesson/779916/3_1_7.csv')
```

```python
df
```

```python
p1 = df.pivot_table(
    index='user_id',
    columns='car_type',
    values='car_id',
    aggfunc = ['count']
).fillna(0)
```

```python
p1.columns = [f'{i[1]}_{i[0]}' for i in p1.columns]
```

```python
p1.reset_index(inplace=True)
```

```python
p1
```

```python
df
```

```python
p2 = df.pivot_table(
    index='user_id',
    values=['distance', 'ride_cost'],
    aggfunc=['min','max','mean']
).fillna(0)
```

```python
p2.columns = [f'{i[1]}_{i[0]}' for i in p2.columns]
```

```python
p2.reset_index(inplace=True)
```

```python
p2
```

```python
df
```

```python
p3 = df.pivot_table(
    index='user_id',
#     columns='car_type',
    values='car_id',
    aggfunc = ['nunique']
).fillna(0)
```

```python
p3.columns = [f'{i[1]}_{i[0]}' for i in p3.columns]
```

```python
p3.reset_index(inplace=True)
```

```python
p3
```

```python
merged_p = pd.merge(p1, p2, on='user_id')
```

```python
merged_p = pd.merge(merged_p, p3, on='user_id')
```

```python
merged_p
```

```python
def crazy_driver_calculate(x): 
    if (x > 100): 
        return 1
    else:
        return 0    
```

```python
df['crazy_driver'] = df['speed_max'].apply(crazy_driver_calculate)
```

```python
df
```

```python
merged_p = pd.merge(merged_p, df[['user_id','crazy_driver']], on='user_id')
```
